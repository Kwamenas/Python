```python
###Monte Carlo

days=365


dt=1/365


mu=rets['GOOGL'].mean()

sigma=rets['GOOGL'].std()
```

```python
def monte_carlo(start_price,days,mu,sigma):
    price=np.zeros(days)
    start_price=price[0]
    
    shock=np.zeros(days)
    drift=np.zeros(days)
    
    for x in range(1,days):
        shock[x]=np.random.normal(loc=mu * dt, scale=sigma * np.sqrt(dt))
        drift[x]= mu * dt
        
        price[x]=price[x-1] + (price[x-1]* (drift[x]+shock[x]))
        
    return price 
        
        
```

```python
start_price=5005.2

for x in range(100):
    plt.plot(monte_carlo(start_price,days,mu,sigma))
         
plt.xlabel("Days")
         
plt.ylabel("Price")
         
plt.title('Monte Carlo')
```

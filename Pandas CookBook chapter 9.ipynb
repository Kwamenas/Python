```python
def weighted_average(df):
    weight_m = df['UGDS'] * df['SATMTMID']
    weight_v = df['UGDS'] * df['SATVRMID']
    wm_avg = weight_m.sum() / df['UGDS'].sum()
    wv_avg = weight_v.sum() / df['UGDS'].sum()
    data = {'w_math_avg': wm_avg,
           'w_verbal_avg': wv_avg,
           'math_avg': df['SATMTMID'].mean(),
           'verbal_avg': df['SATVRMID'].mean(),
           'count': len(df)
           }
    return pd.Series(data)
```

```python
flight.assign(ON_TIME=flight['ARR_DELAY'].lt(15).astype(int))[['AIRLINE','ORG_AIR','ON_TIME']]
```

```python
def max_streaks(s):
    s1=s.cumsum()
    return (s.mul(s1).diff().where(lambda x:x<0).ffill().add(s1,fill_value=0).max())
    
    
```

```python
(flight
 .assign(ON_TIME=flight['ARR_DELAY'].lt(15).astype(int))
 .sort_values(['MONTH','DAY','SCHED_DEP'])
 .groupby(['AIRLINE','ORG_AIR'])['ON_TIME'].agg(['mean','size',max_streaks])
 .round(2)
)
```

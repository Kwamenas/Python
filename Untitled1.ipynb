```python
def pages_to_sentence(*pages):
    """Remember we are just returning a list of sentence from our article"""
    sentence=[]
    
    for page in pages:
        p=wikipedia.page(page)
        doc=nlp(p.content)
        #sentence.append([sent.text for sent in doc.sents])
        sentence+=[sent.text for sent in doc.sents]
    
    return sentence
    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

###my X and y 
documents=animal_sents+language_sents
labels=['animal']*len(animal_sents) + ['language']*len(language_sents)


###we will use stop words so lets check our stop words
STOP_WORDS_str=" ".join(STOP_WORDS)
STOP_WORD_LEM=set(word.lemma_.lower() for word in nlp(STOP_WORDS_str))

##Now my model
TfidVect= TfidfVectorizer(stop_words=STOP_WORD_LEM,tokenizer=lemmatizer,ngram_range=(1,2))
pipe=Pipeline([('vector',TfidVect),('classifier',MultinomialNB())])

pipe.fit(documents,labels)
```

```python
def page_to_sentence(*pages):
    sentence=[]
    for page in pages:
        p=wikipedia.page(page)
        doc=nlp(p.content)
        sentence +=[sent.text for sent in doc.sents]

    return sentence
```

```python

corpus=page_to_sentence('Amazon_rainforest','Amazons','Amazon_(company)')
```

```python
labels=['forest']*len(corpus) + ['Company']*len(corpus) +['Tribe']*len(corpus)

```

```python
def get_corpus(words):
    corpus=[]
    label=[]
    
    for k,v in words.items():
        doc=page_to_sentence(*v)
        corpus+=doc
        label+=[k]*len(corpus)
    
    return corpus,label
        
        
        
        
words={'forest':['Amazon_rainforest'],'Company':['Amazon_(company)'],'Tribe':['Amazons']}
```

```python
corpus,labels=get_corpus(words)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test=train_test_split(documents,labels,test_size=0.2,random_state=0)
```

```python
param_grid=[{'tokenizer':[None,lemmatizer],'ngram_range':[(1,1),(1,2)]}]
pipeg=Pipeline([('vectorizer',GridSearchCV(TfidfVectorizer(),param_grid,cv=5,verbose=1,scoring='neg_mean_squared_error')),
              ('classifier',MultinomialNB())])
pipeg.fit(X_train,y_train)
```
